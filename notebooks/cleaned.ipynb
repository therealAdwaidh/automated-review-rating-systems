```python
import pandas as pd
df = pd.read_csv('../data/cleaned_dataset/cleaned_data.csv')
df.tail()
```

```python
# Check available scores
print("Unique Score values:", df['Score'].unique())
print(df['Score'].value_counts())
```

Taking more samples until balanced dataset becomes 2000+ samples per Score.

```python
# Step 1: Take first 15,000 samples as balanced dataset
balanced_df = df.iloc[:100000]
# Step 2: Take the remaining samples as imbalanced dataset
imbalanced_df = df.iloc[100000:]
```

```python
# Step 2: Save the balanced dataset
balanced_df.to_csv("../data/balanced_dataset/balanced.csv", index=False)
print("✅ Saved balanced.csv with shape:", balanced_df.shape)
```

```python
# Step 3: Remove balanced samples from the main dataset to create imbalanced
remaining_df = df.drop(balanced_df.index)
remaining_df.to_csv("../data/imbalanced_dataset/imbalanced.csv", index=False)
print("✅ Saved imbalanced.csv with shape:", remaining_df.shape)
```
