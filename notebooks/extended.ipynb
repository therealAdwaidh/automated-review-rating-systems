```python
import pandas as pd
df = pd.read_csv('../data/extended_dataset/Reviews.csv')
df.head()
```

View & Inspect the Data

```python
# Preview first rows
print(df.head())

# Check last rows
print(df.tail())

# Data summary (column names, types, missing values)
print(df.info())

# Summary statistics for numeric columns
print(df.describe())
```

Handle Missing Data

```python
print(df.isnull().sum())
```

View datas

```python
df = df[['Score', 'Summary']]
df.head()
```

Drop Unnecessary Columns/Rows

```python
df = df.dropna(subset=['Score', 'Summary'])
```

Remove Duplicates

```python
df = df.drop_duplicates()
```

View missing datas

```python
print(df.isnull().sum())
```

```python
import re

def clean_text(text):
    text = str(text).lower()                    # lowercase
    text = re.sub(r'http\S+', '', text)        # remove URLs
    text = re.sub(r'[^a-z0-9\s]', '', text)    # remove special characters
    text = re.sub(r'\s+', ' ', text).strip()   # remove extra spaces
    return text
```

```python
df['Summary'] = df['Summary'].apply(clean_text)
```

remove columns with less than 3 words

```python
df = df[df['Summary'].apply(lambda x: len(x.split()) >= 3)]
```

```python
df.to_csv("../data/cleaned_dataset/cleaned_data.csv", index=False)
```
