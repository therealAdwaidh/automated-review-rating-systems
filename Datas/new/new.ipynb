```python
import pandas as pd

# Load datasets
balanced_df = pd.read_csv('balanced.csv')
imbalanced_df = pd.read_csv('imbalanced.csv')

# Step 1: take first 10,000 samples from imbalanced dataset
subset = imbalanced_df[['Summary', 'Score']].iloc[:40000]

# Step 2: add them to balanced dataset
combined_df = pd.concat([balanced_df, subset], ignore_index=True)

# Step 3: remove those 10,000 samples from imbalanced dataset
updated_imbalanced_df = imbalanced_df.iloc[40000:].reset_index(drop=True)

# Step 4: save results
combined_df.to_csv('combined_dataset.csv', index=False)
updated_imbalanced_df.to_csv('imbalanced_updated.csv', index=False)
```
